```python
from lib.Browser import BROWSER
import time
import re
```

```python
place='USA'
Browser=BROWSER(headless=False)
Browser.search(f"https://nominatim.openstreetmap.org/ui/search.html?q={place}")
# You should now see a new browser emulator (selenium) to come up if headless is false
```

```python
Browser.jquery() 
Id = None
while Id is None:
    Id=Browser("""return $('div#searchresults').children().first().find('a').attr('href')""")
    time.sleep(1)

number=re.findall(r'osmid=(\d+)',Id)[0]
number
```

```python
Browser.search(f'http://polygons.openstreetmap.fr/get_geojson.py?id={number}&params=0')
Browser.jquery()
json=Browser("""return $('html').text()""")
json=eval(json)
json
```
